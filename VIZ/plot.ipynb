```python
# Assuming you have the dataframe named rand_dsa
# Let's extract the relevant columns: 'Time' and 'LEO1_Melbourne_SNR_dB'
time = rand_dsa['Time']
leo1_melbourne_snr = rand_dsa['LEO1_Melbourne_SNR_dB']

# Plotting the data
plt.figure(figsize=(10,6))
plt.plot(time, leo1_melbourne_snr, label='LEO1 Melbourne SNR', color='b')
plt.xlabel('Time')
plt.ylabel('LEO1 Melbourne SNR (dB)')
plt.title('LEO1 Melbourne SNR vs Time')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
